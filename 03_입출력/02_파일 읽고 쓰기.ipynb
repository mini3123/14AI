```python
# 파일 생성하기
f = open("새파일.txt","w")
f.close()
```

```python
f = open("새 파일2.txt","w",encoding="UTF-8")
for i in range(1,11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
f = open("./새 파일2.txt","r",encoding="UTF-8")
line = f.readline()
print(line)
f.close()
```

```python
f = open("./새 파일2.txt","r",encoding="UTF-8")
while True:
    line = f.readline()
    if not line: break
    print(line)
f.close()
```

```python
# readlines()
f = open("./새 파일2.txt","r",encoding="UTF-8")
lines = f.readlines()
print(lines)
f.close()
```

```python
# 파일에 새로운 내용 추가하기
f = open("./새 파일2.txt","a",encoding="UTF-8")
for i in range(22,31):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
f = open("./새 파일2.txt","r",encoding="UTF-8")
while True:
    line = f.readline()
    if not line: break
    print(line)
f.close()
```

```python
# with 문과 함께 사용하기
with open("foo.txt","w") as f:
    f.write("Life is too short, you need Python")
```
