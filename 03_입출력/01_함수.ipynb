```python
n = int(input("숫자 입력: "))

if n < 2:
    print("2 이상 입력")
else:

    is_prime = True

    for i in range(2,n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(n,"은 소수입니다")
    else:
        print(n,"은 소수가 아닙니다")
```

```python
num = int(input("숫자를 입력하세요: "))

if num < 2:
    print("2 이상의 숫자만 입력 가능합니다.")
else:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    
    if is_prime:
        print(f"{num}은(는) 소수입니다.")
    else:
        print(f"{num}은(는) 소수가 아닙니다.")

    prime_list = []
    for i in range(2, num + 1):
        check = True
        for j in range(2, int(i**0.5) + 1):
            if i % j == 0:
                check = False
                break
        if check:
            prime_list.append(i)

    print(f"\n1부터 {num}까지의 소수:")
    print(", ".join(map(str, prime_list)))
    print(f"총 {len(prime_list)}개의 소수가 있습니다.")
```

```python
money = 10000 

print("=== ATM 시스템 ===")

while True:
    print("\n1. 잔액 조회")
    print("2. 입금")
    print("3. 출금")
    print("4. 종료")
    
    choice = input("선택: ")

    if choice == '1':
        print(f"현재 잔액: {money:,}원")

    elif choice == '2':
        amount = int(input("입금액: "))
        if amount > 0:
            money += amount
            print(f"입금 완료! 현재 잔액: {money:,}원")
        else:
            print("올바른 금액을 입력해주세요.")

    elif choice == '3':
        amount = int(input("출금액: "))
        if amount > money:
            print(f"잔액이 부족합니다. 현재 잔액: {money:,}원")
        elif amount <= 0:
            print("올바른 금액을 입력해주세요.")
        else:
            money -= amount
            print(f"출금 완료! 현재 잔액: {money:,}원")

    elif choice == '4':
        print("이용해 주셔서 감사합니다.")
        break

    else:
        print("잘못된 선택입니다. 1~4번 메뉴를 선택해주세요.")
```

```python
import random

number = random.randint(1, 100)
count = 0  

print("1부터 100 사이의 숫자를 맞춰보세요!")

while True:
    count += 1
  
    try:
        user = int(input(f"시도 {count}: "))
    except ValueError:
        print("숫자만 입력해 주세요!")
        count -= 1  
        continue

    if user < number:
        print("Up!")
    elif user > number:
        print("Down!")
    else:
        print(f"정답입니다! {count}번 만에 맞추셨습니다!")
        break
```

```python
num_students = int(input("학생 수를 입력하세요: "))
students = []

for i in range(1, num_students + 1):
    print(f"\n[{i}번 학생]")
    name = input("이름: ")
    kor = int(input("국어: "))
    eng = int(input("영어: "))
    math = int(input("수학: "))
    
    total = kor + eng + math    
    avg = total / 3
    
    if avg >= 90: 
        grade = 'A'
    elif avg >= 80: 
        grade = 'B'
    elif avg >= 70: 
        grade = 'C'
    elif avg >= 60: 
        grade = 'D'
    else: 
        grade = 'F'
    
    student_info = {
        "이름": name,
        "국어": kor,
        "영어": eng,
        "수학": math,
        "평균": avg,
        "등급": grade
    }
    students.append(student_info)
    
    print(f"평균: {avg:.2f}점 (등급: {grade})")

best_student = students[0]
worst_student = students[0]
total_kor = 0
total_eng = 0
total_math = 0

for s in students:
    
    if s["평균"] > best_student["평균"]:
        best_student = s
    if s["평균"] < worst_student["평균"]:
        worst_student = s
    
    total_kor += s["국어"]
    total_eng += s["영어"]
    total_math += s["수학"]

print("\n=== 전체 통계 ===")
print(f"최고 점수: {best_student['이름']} ({best_student['평균']:.2f}점)")
print(f"최저 점수: {worst_student['이름']} ({worst_student['평균']:.2f}점)")
print(f"국어 평균: {total_kor / num_students:.2f}점")
print(f"영어 평균: {total_eng / num_students:.2f}점")
print(f"수학 평균: {total_math / num_students:.2f}점")
```

```python
products = {
    "1": {"name": "티셔츠", "price": 15000},
    "2": {"name": "청바지", "price": 30000},
    "3": {"name": "두쫀쿠", "price": 5000},
    "4": {"name": "모자", "price": 12000},
    "5": {"name": "후드집업", "price": 45000}
}

cart = []

while True:
    print("\n=== 쇼핑몰 메뉴 ===")
    print("1. 상품목록  2. 장바구니 담기  3. 장바구니 보기  4. 결제  5. 종료")
    choice = input("선택: ")

    if choice == '1':
        print("\n--- 판매 상품 ---")
        for key, value in products.items():
            print(f"{key}. {value['name']} - {value['price']:,}원")

    elif choice == '2':
        p_num = input("담을 상품 번호를 입력하세요: ")
        if p_num in products:
            cart.append(products[p_num])
            print(f"'{products[p_num]['name']}'이(가) 장바구니에 담겼습니다.")
        else:
            print("존재하지 않는 상품 번호입니다.")

    elif choice == '3':
        if not cart:
            print("\n장바구니가 비어 있습니다.")
        else:
            print("\n--- 내 장바구니 ---")
            temp_total = 0
            for item in cart:
                print(f"- {item['name']}: {item['price']:,}원")
                temp_total += item['price']
            print(f"현재 합계: {temp_total:,}원")

    elif choice == '4':
        if not cart:
            print("결제할 상품이 없습니다.")
            continue
        
        total_price = 0
        for item in cart:
            total_price += item['price']
            
        delivery_fee = 5000 if total_price < 30000 else 0
        
        print(f"\n총 상품 금액: {total_price:,}원")
        
        use_coupon = input("10% 할인 쿠폰을 사용하시겠습니까? (네/아니오): ").lower()
        if use_coupon == '네':
            discount = total_price * 0.1
            total_price -= discount
            print(f"할인 금액: -{int(discount):,}원")
        
        final_price = total_price + delivery_fee
        print(f"배송비: {delivery_fee:,}원 (3만원 이상 무료)")
        print(f"--- 최종 결제 금액: {int(final_price):,}원 ---")
        
        cart = [] 
        print("결제가 완료되었습니다. 감사합니다!")

    elif choice == '5':
        print("프로그램을 종료합니다.")
        break
```

```python
# 입력값이 몇 개가 될지 모를 때는 어떻게 할까?
def add_many(*args):
    result = 0
    for i in args:
        result += i
    return result

result = add_many(1,2,3,4,5,6,7)
print(result)
```

```python
def add_mul(choice,*agrs):
    if choice == "add":
        result = 0
        for i in agrs:
            result = result + i
    elif choice == "mul":
        result = 1
        for i in agrs:
            result = result * i
    return result
```

```python
result = add_mul("add",1,2,3,4,5)
print(result)
```

```python
# lambda 예약어
# 함수 이름 = lammda 매개변수1 , 매개변수2 , ... : (수행할 문장)
def add(a,b):
    return a + b

add = lambda a,b : a + b
```

```python
# 함수의 독스트링(Docstring)
def add(a,b):
    """
    두 숫자를 더하는 함수
    
    Parameters:
    a(int,float): 첫 번째 숫자
    b(int,float): 두 번째 숫자

    returns:
    int,float: 두 숫자의 합이다.

    """
    return a + b
print(add.__doc__)
```
