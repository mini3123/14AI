```python
# 튜플은 어떻게 만들까?

a = ()
b = (1,) # 하나의 요소만 콤마를 붙임
c = (1,2,3)
d = 1,2,3 # 괄호 생략 가능
e = ("a","b",("c","d")) # 튜플 안에 튜플 가능
f = 1,
g = 1
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
```

```python
# 튜플 요소값을 삭제하려 할 때

a = (1,2,"a","b")
del a[0]
print(a)
```

```python
# 튜플 요소값을 변경하려 할 때

a = (1,2,"a","b")
a[0] = "C"
```

```python
a = (1,2)
a = a * 2
print(a)
```

```python
a = 10 # 변경가능한 타입 아님 리스트는 변경 가능
print(id(a))
a = 20
print(id(a))
```

```python
a = [1,2,3]
print(id(a))

a.append(4)
print(a)
print(id(a))
```

```python
# 튜플 인덱싱 하기
a = (1,2,3,4)
print(a[0])
```

```python
# 튜플 슬라이싱 하기
a = (1,2,3,4)
print(a[:2])
```

```python
# 튜플에 더하기
a = (1,2)
print(id(a))
b = (3,4)
print(id(b))
c = a + b
print(id(c))
print(c)
```

```python
# 튜플에 곱하기
a = (3,4)
b = a*3
print(b)
```

```python
# 튜플 길이 구하기
a = (1,2,3,4)
len(a)
```
