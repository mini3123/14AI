```python
# 딕셔너리(key+value)는 어떻게 만들까?
dic = {"name" : "ken","phone":"010-4123-2131","birth":"01-01"}
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기
a = {1:"a"}
a[2]="b"
print(a)
a["name"] = "ken"
print(a)
a[(1,2)] = "hobby"
print(a)

a[[3,4]] = "ms" #변경이 가능한 것만 가능
print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1] #key값
print(a)
del a["name"]
print(a)
```

```python
# 딕셔너리에서 key를 사용해 value 얻기
profile = {
  "name" : "홍길동",
  "age" : 25
}

profile["name"]
profile["age"]
```

```python
# 딕셔너리에 관련된 함수
# 1. key 리스트 만들기 - keys
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
print(profile.keys())

# 2. value 리스트 만들기 values
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
print(profile.values())

# 3. key:values 쌍 얻기 items()
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
print(profile.items())

# 4. key, value 쌍 모두 지우기 - clear
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
print(profile.clear())

# 5. key로 value 얻기 - get
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
print(profile.get("이름"))

# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
"이름" in profile

# key로 values얻기 - pop
profile = {
  "이름" : "홍길동",
  "나이" : 25,
  "번호" : "010-1231-1232"
}
temp = profile.pop("이름") # 빠져나옴
print(temp)
print(profile)
```
