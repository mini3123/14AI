```python
print(1,bool(1))
print(0,bool(0))
print(2,bool(2))
print("Python",bool("Python"))
print("",bool(" "))
print("",bool("")) # 쌍따옴표가 붙어있으면 False 그 반대는 True
print("[]",bool([]))
print("[1,2,3]",bool([1,2,3])) # 리스트 안에 값이 있으면 True 없으면 False

```

```python
# 논리 연산자
# 1. and 연산자

print(True and True)
print(True and False)

# 2. or 연산자

print(True or True)
print(True or False)
print(False or False)

# 3. not 연산자

print(not True)
print(not False)
print(not 1)
print(not 0)


```
