```python
students = [
 {"이름":"김민수","국어":85,"영어":90,"수학":78},
 {"이름":"이영희","국어":92,"영어":88,"수학":95},
 {"이름":"박철수","국어":78,"영어":85,"수학":82}
]

total1 = students[0]["국어"] + students[0]["영어"] + students[0]["수학"]
avg1 = total1 / 3

print("이름:", students[0]["이름"])
print("국어:", students[0]["국어"], ", 영어:", students[0]["영어"], ", 수학:", students[0]["수학"])
print("총점:", total1 ,", 평균:", round(avg1,2) ),
print("---")
```

```python
text = "python is great python is fun python is powerful"
words = text.split()

count = {}

for w in words:
    if w in count:
        count[w] += 1
    else:
        count[w] = 1

for k,v in count.items():
    print(k,":",v,"번")
```
