```python
# 큰 따옴표
"Life is too short, you need Python"

# 작은 따옴표 둘러싸기
'Python is fun'

# 큰 따옴표 3개를 연속으로 둘러싸기
"""Life is too short, you need Python"""


# 작은 따옴표 3개를 연속으로 둘러싸기
'''Life is too short, you need Python'''
```

```python
food = "Python's favorite food is perl"
say = '"Python is very easy. " he says' # 큰 따옴표 밖에 작은 따옴표를 넣으면 하나의 문자열로 인식함

# 역슬래시를 사용해서 작은따옴표와 큰따옴표를 문자열에 포함하기
food = 'Python\'s favorite food is perl'
food

say = "\"Python is very easy. \" he say"
```

```python
# 줄을 바꾸기 위한 이스케이프 코드 \n = 줄바꿈
python = "Life is too short,\n you need Python"
print(python)
```
