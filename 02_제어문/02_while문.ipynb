```python
treehit = 0
while treehit < 10:
    treehit = treehit + 1
    print(f"나무를 {treehit}번 찍었습니다")
    if treehit == 10:
      print("나무가 넘어갑니다")
      
```

```python
number = 0
prompt = """
1. Add
2. Del
3. List
4. Quit
"""
while number != 4:
    print(prompt)
    number = int(input("1~4까지 숫자 중 하나를 고르세요."))
    if number == 4:
        print("종료되었습니다.")
```

```python
# while문 강제로 빠져나가기

coffee = 5
cost = 300
while cost:
    print("돈을 받았으니 커피를 줍니다.")
    coffee -= 1
    print(f"남은 커피는 {coffee}개 남았습니다.")
    if coffee == 0:
        print("남은 커피가 없습니다.")
        break    
```

```python
distance = float(input("거리를 입력하세요 (km): "))
hour = int(input("현재 시간을 입력하세요 (0-23): "))
fare = 3000

if distance > 2:
    fare += (distance - 2) * 1000

if hour >= 22 or hour <= 4:
    fare *= 1.2

print(f"요금: {int(fare)}원")
```

```python
coffee = 10

while True:
    money = int(input("돈을 넣어주세요 : "))
    if money == 300:
        print("커피를 줍니다.")
        coffee -=1
    elif money > 300:
        print(f"거스름돈 {money - 300} 을 주고, 커피를 줍니다.")
        coffee -=1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print(f"현재 남은 커피의 갯수는 {coffee}개 입니다.")
    if coffee == 0:
        print("판매를 중지합니다.")
        break
```

```python
# while문의 맨 처음으로 돌아가기 - continue
# 1~10 중에서 홀수만 출력하기
n = 0

while n < 10:
    n += 1
    if n % 2 == 0:
      continue
    print(n)
```
