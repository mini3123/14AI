```python
# while문 기본 구조

# while 조건문:
#     수행할문장1
#     수행할문장2
#     수행할문장3
#     수행할문장4
#     ...
```

```python
treehit = 0
while treehit < 10:
    treehit = treehit + 1
    print(f"나무를 {treehit}번 찍었습니다")
    if treehit == 10:
      print("나무가 넘어갑니다")
      
```

```python
number = 0
prompt = """
1. Add
2. Del
3. List
4. Quit
"""
while number != 4:
    print(prompt)
    number = int(input("1~4까지 숫자 중 하나를 고르세요."))
    if number == 4:
        print("종료되었습니다.")
```

```python
# while문 강제로 빠져나가기

coffee = 5
cost = 300
while cost:
    print("돈을 받았으니 커피를 줍니다.")
    coffee -= 1
    print(f"남은 커피는 {coffee}개 남았습니다.")
    if coffee == 0:
        print("남은 커피가 없습니다.")
        break    
```
