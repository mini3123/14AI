```python
# for문의 기본구조
# for 변수 in 리스트(또는 튜플, 문자열 ):
#     수행할 문장1
#     수행할 문장2
#     ...

num_list = ["one","two","three"]
for num in num_list:
    print(num)
```

```python
a = [(1,2),(3,4),(5,6)]
for i in a:
    print(i)
```

```python
tuple_list = [(1,2),(3,4),(5,6)]
for a in tuple_list:
    print(a)
```

```python
a , b = 1, 2
[a,b] = [1,2]
[a,b] = (1,2,3)
```

```python
marks = [90,25,67,45,80]
n = 0

for mark in marks:
    n += 1
    if mark > 60:
      print(f"{n}번째 학생은 통과입니다.")
    else:
      print(f"{n}번째 학생은 탈락입니다.")
```

```python
# for문과 continue문
marks = [90,25,67,45,80]
n = 0

for mark in marks:
    n += 1
    if mark < 60:
        continue
    print(f"{n}번째 학생은 합격입니다.")
```

```python
result = 0
for num in range(1,10+1):
    result += num
print(result)
```

```python
marks = [90,25,67,45,80]

for n in range(len(marks)):
    if marks[n] < 60:
        continue
    print(f"{n+1} 번째 학생은 합격입니다.")
  
```

```python
# for와 range를 이용한 구구단
for i in range(2,10):
    for j in range(1,10):
        print(f"{i} X {j} = ",i*j,end=" ")
```

```python
for i in range(2,10):
    for j in range(1,10):
        print(f"{i} X {j} = ",i*j)
    print("")
```

```python
for i in range(1,n+1):
    print(" "*(n-i)+"*"*(2*i-1))
for i in range(n-1,0,-1):
    print(" "*(n-i)+"*"*(2*i-1))
```

```python
# 리스트 컴프리헨션 사용하기
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)
print(result)
```

```python
a = [1,2,3,4]
result = [num*3 for num in a]
print(result)
```

```python
# for문과 break문 사용하기
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# enumerate() 함수 활용하기
fruits = ["apple","banana","orange"]

for i in fruits:
    print(f"번째 : {i}")
```

```python
fruits = ["apple","banana","orange"]

for i,fruit in enumerate(fruits):
    print(f"{i+1}번째 : {fruit}")
```

```python
# zip 함수로 여러 리스트 함께 순회하기
names = ["홍길동","강호동","유재석"]
scores = [60,70,80]

for name,score in zip(names,scores):
    print(f"{name}의 점수는 {score}이다.")
```

```python
# 3개 이상의 리스트 zip()
names = ["홍길동","강호동","유재석"]
scores = [60,70,80]
Korean = [30,40,50]

for name,score , korea in zip(names,scores,Korean):
    print(f"{name}의 점수는 {score}이고 국어 점수는 {korea}이다.")
```
