```python
# if문은 왜 필요할까?
# 돈이 있으면 택시를 타고 가고, 돈이 없으면 걸어간다.
# 우유 하나 사와, 달걀 있으면 6개 사와

money = True

if money: #로지컬 테스트
    print("택시를 타고 간다.")
else:
    print("걸어간다")
```

```python
egg = True

if egg:
    print("우유 6개 사와")
else:
    print("우유 하나 사와")
```

```python
# if문의 기본 구조

# if 조건문:
#     수행할 문장1
#     수행할 문장2
#     ...
# else:
#     수행할 문장1
#     수행할 문장2
#     ...
```

```python
money = True

if money:
    print("택시를")
    print("타고")
    print("간다")
else:
    print("걸어서")
    print("간다")

```

```python
# x < y  작다
# x > y  크다
# x >= y 크거나 같다
# x <= y 작거나 같다
# x == y 같다
# x != y 같지 않다
```

```python
# 돈이 30000원 이상이거나 카드가 있다면 택시를 타고 가고 그렇지 않으면 걸어서 간다
money = 30000
card = True

if money >= 30000 or card == True:
    print("택시를 탄다.")
else:
    print("걸어서 간다.")

money = 30000
card = False

if money >= 30000 and card == True:
    print("택시를 탄다.")
else:
    print("걸어서 간다.")    
```

```python
# 주머니에 돈이 있으면 택시를 타고 가고 없으면 걸어가라
pocket = ["paper","phone","money"]

if "money" in pocket:
    print("택시를 타고가라")
else:
    print("걸어서 가라")
```

```python
# 다양한 조건을 판단하는 elif

# 주머니에 돈이 있으면 택시를 타고 가고 주머니에 돈이 있지만 카드가 있으면 택시를 타고 가고
# 돈도 없고 카드도 없으면 걸어가라
pocket = ["phone"]
card = False

if "money" in pocket:
    print("택시를 타고 가라")
elif card:
    print("택시를 타고 가라")
else:
    print("걸어서 가라")
```
