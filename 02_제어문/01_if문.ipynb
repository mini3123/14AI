```python
# if문은 왜 필요할까?
# 돈이 있으면 택시를 타고 가고, 돈이 없으면 걸어간다.
# 우유 하나 사와, 달걀 있으면 6개 사와

money = True

if money: #로지컬 테스트
    print("택시를 타고 간다.")
else:
    print("걸어간다")
```

```python
# x < y  작다
# x > y  크다
# x >= y 크거나 같다
# x <= y 작거나 같다
# x == y 같다
# x != y 같지 않다
```

```python
# 돈이 30000원 이상이거나 카드가 있다면 택시를 타고 가고 그렇지 않으면 걸어서 간다
money = 30000
card = True

if money >= 30000 or card == True:
    print("택시를 탄다.")
else:
    print("걸어서 간다.")

money = 30000
card = False

if money >= 30000 and card == True:
    print("택시를 탄다.")
else:
    print("걸어서 간다.")    
```

```python
# 주머니에 돈이 있으면 택시를 타고 가고 없으면 걸어가라
pocket = ["paper","phone","money"]

if "money" in pocket:
    print("택시를 타고가라")
else:
    print("걸어서 가라")
```

```python
# 다양한 조건을 판단하는 elif

# 주머니에 돈이 있으면 택시를 타고 가고 주머니에 돈이 있지만 카드가 있으면 택시를 타고 가고
# 돈도 없고 카드도 없으면 걸어가라
pocket = ["phone"]
card = False

if "money" in pocket:
    print("택시를 타고 가라")
elif card:
    print("택시를 타고 가라")
else:
    print("걸어서 가라")
```

```python
# 성적이 90점 이상이면 "A학점"
# 성적이 90점 미만이면서 80점 이상이면 "B학점"
# 성적이 80점 미만이면서 70점 이상이면 "C학점"
# 성적이 70점 미만이면서 60점 이상이면 "D학점"
# 그 외는 "F학점"

grade = int(input("성적을 입력하세요: "))

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
elif grade >= 60:
    print("D")
else:
    print("F")
```

```python
# match class 구문

# match 변수
#  case 패턴 1:
#   수행할 무장 
#  case 패턴 2:
#   수행할 문장

grade = input("A~F중 하나의 값을 입력하시오. : ")

match grade:
    case "A" | "B" | "C" | "D" | "E":
        print("합격")
    case "F":
        print("탈락")
    case _:
        print("잘못 입력하셨습니디. 다시 시도해주세요.")
```
