```python
class FourCal:
    def setdata(self,first,second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def div(self):
        result = self.first // self.second
        return result
    
cal1 = FourCal()
cal1.setdata(7,5)
print(cal1.add())
print(cal1.sub())
print(cal1.mul())
print(cal1.div())
```

```python
class Product:
    """상품 정보를 관리하는 클래스"""
    
    def __init__(self, product_id, name, price, weight, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
    
    def get_info(self):
        """상품 정보를 딕셔너리로 반환"""
        return {
            "ID": self.product_id,
            "Name": self.name,
            "Price": self.price,
            "Weight": self.weight,
            "Stock": self.stock
        }
    
    def update_stock(self, quantity):
        """재고 수량 업데이트 (부족하면 False, 성공하면 True)"""
        if self.stock >= quantity:
            self.stock -= quantity
            return True
        return False
    
    def apply_discount(self, discount_rate):
        """할인율(%)을 적용한 가격 반환 (원본 가격 유지)"""
        discount_amount = self.price * (discount_rate / 100)
        return int(self.price - discount_amount)


class ShoppingCart:
    """장바구니를 관리하는 클래스"""
    
    def __init__(self):
        self.items = [] # 상품 정보를 저장할 리스트
    
    def add_product(self, product, quantity):
        """장바구니에 상품 객체와 수량 추가"""
        self.items.append({'product': product, 'quantity': quantity})
    
    def remove_product(self, product_id):
        """상품 ID로 장바구니에서 제거"""
        self.items = [item for item in self.items if item['product'].product_id != product_id]
    
    def get_total_price(self):
        """모든 상품의 총 금액 계산"""
        total = 0
        for item in self.items:
            total += item['product'].price * item['quantity']
        return total
    
    def get_total_weight(self):
        """모든 상품의 총 무게 계산"""
        total_w = 0
        for item in self.items:
            total_w += item['product'].weight * item['quantity']
        return total_w
    
    def show_cart(self):
        """장바구니 내용을 보기 좋게 출력"""
        print(f"\n{'='*10} 장바구니 목록 {'='*10}")
        for item in self.items:
            p = item['product']
            print(f"- {p.name} | 수량: {item['quantity']}개 | 가격: {p.price * item['quantity']:,}원")
        print(f"{'='*34}")


class Order:
    """주문 정보를 관리하는 클래스"""
    
    def __init__(self, order_id, cart, customer_name):
        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name
        self.shipping_fee = 3000 # 기본 배송비
    
    def calculate_final_price(self):
        """최종 결제 금액 계산 (상품 총액 + 배송비)"""
        total_price = self.cart.get_total_price()
        # 예시 로직: 10만원 이상 구매 시 배송비 무료
        if total_price >= 100000:
            self.shipping_fee = 0
        return total_price + self.shipping_fee
    
    def print_receipt(self):
        """주문서 출력"""
        print(f"\n{'#'*10} 주문 영수증 {'#'*10}")
        print(f"주문 번호: {self.order_id}")
        print(f"고객 성함: {self.customer_name}")
        print("-" * 30)
        
        # 장바구니 품목 출력
        for item in self.cart.items:
            p = item['product']
            print(f"{p.name:10} x {item['quantity']} : {p.price * item['quantity']:,}원")
            
        final_price = self.calculate_final_price()
        print("-" * 30)
        print(f"상품 합계: {self.cart.get_total_price():,}원")
        print(f"배송 요금: {self.shipping_fee:,}원")
        print(f"최종 결제: {final_price:,}원")
        print(f"{'#'*32}\n")

# --- 실행 코드 ---

# 1. 상품 생성
laptop = Product("P001", "노트북", 1200000, 2.5, 10)
mouse = Product("P002", "무선마우스", 35000, 0.2, 50)
keyboard = Product("P003", "기계식키보드", 89000, 1.0, 30)

# 2. 장바구니 생성 및 상품 추가
cart = ShoppingCart()
cart.add_product(laptop, 1)
cart.add_product(mouse, 2)
cart.add_product(keyboard, 1)

# 3. 장바구니 확인
cart.show_cart()

# 4. 주문 생성
order = Order("ORD20260211001", cart, "홍길동")

# 5. 주문서 출력
order.print_receipt()
```
